```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sahus\OneDrive\Desktop\netflix_titles.csv')
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df['type'].value_counts()        # Movies vs TV Shows
df['rating'].value_counts()      # Most common ratings
df['country'].value_counts().head(10)  # Top 10 countries
```

```python
df['type'].value_counts()
```

```python
df['rating'].value_counts()
```

```python
# Movies vs TV Shows
sns.countplot(x='type', data=df)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Netflix EDA Dashboard', fontsize=16)

# 1. Movies vs TV Shows
sns.countplot(x='type', data=df, ax=axes[0,0], palette='Set1')
axes[0,0].set_title('Movies vs TV Shows')

# 2. Top 10 Countries
top_countries = df['country'].value_counts().head(10)
axes[0,1].bar(top_countries.index, top_countries.values, color='steelblue')
axes[0,1].set_title('Top 10 Countries')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Top Ratings
sns.countplot(x='rating', data=df, ax=axes[1,0], 
              order=df['rating'].value_counts().index, palette='Set2')
axes[1,0].set_title('Content Ratings Distribution')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Content added per year
df['year_added'] = df['date_added'].str.strip().str[-4:]
year_counts = df['year_added'].value_counts().sort_index()
axes[1,1].plot(year_counts.index, year_counts.values, marker='o', color='red')
axes[1,1].set_title('Content Added Per Year')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Key Findings

1. Netflix has more than twice as many Movies (6,126) than TV Shows (2,664).
2. The USA produces the most content by far (2,809 titles), followed by India (972).
3. TV-MA is the most common rating, showing Netflix targets adult audiences.
4. Content additions peaked around 2019-2020 and slowed after that.
