```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sahus\OneDrive\Desktop\netflix_titles.csv')
## Clean it quickly
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.dropna(inplace=True)
```

# Netflix Data Visualization - CodeAlpha Task 3

Visualizing trends and patterns in the Netflix Movies & TV Shows dataset.

```python
# Split genres and count each one
genres = df['listed_in'].str.split(',').explode().str.strip()
top_genres = genres.value_counts().head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='Reds_r')
plt.title('Top 10 Genres on Netflix', fontsize=16)
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

## Chart 1 - Top 10 Genres
International Movies dominate Netflix, followed by Dramas and Comedies.
Netflix clearly has a strong international content strategy.


```python
# Pie chart - Movies vs TV Shows
plt.figure(figsize=(8, 8))
type_counts = df['type'].value_counts()
plt.pie(type_counts.values, 
        labels=type_counts.index, 
        autopct='%1.1f%%',
        colors=['#E50914', '#221F1F'],
        startangle=90,
        explode=(0.05, 0))
plt.title('Movies vs TV Shows on Netflix', fontsize=16)
plt.tight_layout()
plt.show()
```

## Chart 2 - Movies vs TV Shows
Movies make up ~70% of Netflix content, with TV Shows at ~30%.
Netflix is primarily a movie platform but TV Shows are growing.

```python
# Line chart - Content added per year
plt.figure(figsize=(12, 6))
df['year_added'] = df['date_added'].str.strip().str[-4:]
year_counts = df['year_added'].value_counts().sort_index()

plt.plot(year_counts.index, year_counts.values, 
         marker='o', color='#E50914', linewidth=2.5, markersize=8)
plt.fill_between(year_counts.index, year_counts.values, alpha=0.2, color='#E50914')
plt.title('Content Added to Netflix Per Year', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Top 10 countries (excluding Unknown)
plt.figure(figsize=(12, 6))
top_countries = df[df['country'] != 'Unknown']['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, hue=top_countries.index, legend=False, palette='Reds_r')
plt.title('Top 10 Countries Producing Netflix Content', fontsize=16)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

## Chart 5 - Top 10 Directors
Analyzing which directors have produced the most content on Netflix.

```python
# Top 10 directors (excluding Unknown)
plt.figure(figsize=(12, 6))
top_directors = df[df['director'] != 'Unknown']['director'].value_counts().head(10)

sns.barplot(x=top_directors.values, y=top_directors.index, hue=top_directors.index, legend=False, palette='Reds_r')
plt.title('Top 10 Directors on Netflix', fontsize=16)
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.tight_layout()
plt.show()
```

## Findings - Top 10 Directors
A small group of directors dominate Netflix content.
Most top directors are from India and USA, reflecting
the dominance of those two countries in overall content.
